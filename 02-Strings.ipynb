## Cadenas y función `print`
Las cadenas pueden encerrarse entre comillas simples ('...') o comillas dobles ("...") con el mismo resultado. \ se puede utilizar para escapar de las comillas:


```python
print('spam eggs')          # Comillas simples
print('doesn\'t')           # use \' para escapar las comillas...
print("doesn't")            # ...o en su lugar use comillas dobles
print('"Yes," he said.')    #
print("\"Yes,\" he said.")
print('"Isn\'t," she said.')
```

La función `print` traduce caracteres especiales de C

```python
s = '\tFirst line.\nSecond line.'  # \n es el código de nueva línea \t inserta una tabulación
print(s)  # con print(), \n introduce una nueva línea
print(r'\tFirst line.\nSecond line.')  # note la r antes de las comillas
```

## Literales de cadena con varias líneas

```python
print("""\
Usage: thingy [OPTIONS]
     -h                        Display this usage message
     -H hostname               Hostname to connect to
""") 
```

El caracter \ quita la nueva linea inicial.

Las cadenas se pueden concatenar (unir) con el operador +, y repetir con *

```python
3 * ("Re" + 2 * 'n' + 'es ')
```

Dos o más cadenas literales contiguas se concatenan automáticamente.

```python
text = ('Put several strings within parentheses '
         'to have them joined together.')
text
```

Las cadenas se pueden indexar, con el primer carácter con índice 0. No hay un tipo de carácter de separación; un caracter es simplemente una cadena de texto de tamaño uno

```python
word = 'Python @ ENSAI'
print(word[0]) # character in position 0
print(word[5]) # character in position 5
```

Los índices también pueden ser números negativos, para comenzar a contar desde la derecha.

```python
print(word[-1])  # last character
print(word[-2])  # second-last character
```

## Cortar cadenas
- Se omite el primer índice predeterminado a cero,
- El segundo índice omitido tiene como valor predeterminado el tamaño de la cadena que se corta.
- El paso se puede configurar con el tercer índice


```python
print(word[:2])  # carácter desde el principio hasta la posición 2 (excluido)
print(word[4:])  # caracteres desde la posición 4 (incluida) hasta el final
print(word[-2:]) # caracteres desde el penúltimo (incluido) hasta el final
print(word[::-1]) # ¡Esta es la cadena invertida!
```

```python
word[::2]
```

Las cadenas de Python no se pueden cambiar, son inmutables. Si necesita una cadena diferente, debe crear una nueva o usar Listas.

```python
import sys
try:
    word[0] = 'J'
except:
    print(sys.exc_info()[0])
```

```python
## Algunos métodos de cadena
print(word.startswith('P'))
```

```python
print(*("\n"+w for w in dir(word) if not w.startswith('_')) )
```
